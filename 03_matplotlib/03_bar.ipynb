## Matplotlib Tutorial 4 - Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np

company = ["GOOGL", "AMZN", "MSFT", "FB"]
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
# uneeded
xpos = np.arange(len(company))
# xpos
plt.xticks(xpos, company)
plt.xlabel("Company")
plt.ylabel("Revenue (bln)")
plt.title("US Tech Stocks")
# plt.bar(xpos, revenue, label="Revenue")
# plt.bar(xpos, profit, label="Profit") # over lap
plt.bar(xpos - 0.2, revenue, width=0.4, label="Revenue")  # sub/add value off xpos
plt.bar(xpos + 0.2, profit, width=0.4, label="Profit")  # side by side
plt.legend()

# works
# plt.bar(company, revenue)
```

```python
plt.yticks(xpos, company)

plt.barh(xpos - 0.2, revenue, label="Revenue")  # sub/add value off xpos
plt.barh(xpos + 0.2, profit, label="Profit")  # side by side
```

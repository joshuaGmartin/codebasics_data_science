```python
import matplotlib.pyplot as plt

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
# plt.hist(blood_sugar)
# plt.hist(blood_sugar, rwidth=0.8)  # by default number of bins/buckets is set to 10
# plt.hist(blood_sugar, rwidth=0.8, bins=4)
# plt.hist(blood_sugar, rwidth=0.95, bins=[80,100,125, 200]) # set bins to known diabetic levels
plt.hist(blood_sugar, rwidth=0.95, bins=[80, 100, 125, 200], color="g", histtype="step")
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel("Blood Sugar")
plt.ylabel("Number of Patients")
plt.title("Men/Women Blood Sugar Analysis")

plt.hist(
    [blood_sugar_men, blood_sugar_women],
    bins=[80, 100, 125, 200],
    rwidth=0.95,
    color=["green", "blue"],
    label=["Men", "Women"],
    orientation="horizontal",
)

plt.legend()
```

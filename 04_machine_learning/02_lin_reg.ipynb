## Machine Learning Tutorial Python - 2: Linear Regression Single Variable

- minimize the difference between line and all data points for best line


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("data/homeprices.csv")
df
```

```python
plt.xlabel("area")
plt.ylabel("price $")
plt.scatter(df.area, df.price)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)  # requires 2D array (Dataframe) for x
# df["area"]
# df[["area"]]
# type(df["area"]) # series
# type(df[["area"]]) # dataframe
# df["area"].shape # (5,)
# df[["area"]].shape (5,1)
```

```python
# reg.predict([[3300]]) # same, needs 2D
reg.predict([[5000]])  # same, needs 2D
# y = mx + b
# reg.coef_ # m
# reg.intercept_ # b
```

```python
df_a = pd.read_csv("data/areas.csv")
df_a.head()
```

```python
prices = reg.predict(df_a)
```

```python
df_a["prices"] = prices
df_a
```

```python
df_a.to_csv("data/prediction.csv", index=False)
```

```python
plt.xlabel("area", fontsize=20)
plt.ylabel("price $", fontsize=20)
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[["area"]]), color="b")
```

```python
reg.predict(df[["area"]])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("data/canada_per_capita_income.csv")
df.head()
```

```python
plt.xlabel("year")
plt.ylabel("per capita income $")
plt.scatter(df.year, df["per capita income (US$)"], marker="+")

reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
# reg.predict([[2020]])
# years = np.arange(1970, 2021)
years = np.arange(1970, 2021).reshape(-1, 1)
# years
plt.plot(years, reg.predict(years), color="r")
```

```python
reg.predict([[2020]])
```

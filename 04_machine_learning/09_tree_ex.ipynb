Titanic data set; ignore red (passenger, name, sibsp, parch, ticket, cabin, embarked) and predict survival rate


```python
import pandas as pd

df = pd.read_csv("data/titanic.csv")
df = df.drop(
    ["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"],
    axis="columns",
)  # use inplace=True
df
```

```python
# split
inputs = df.drop("Survived", axis="columns")
target = df.Survived
```

```python
# clean
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
from sklearn.preprocessing import LabelEncoder

le_Sex = LabelEncoder()

inputs["sex_n"] = le_Sex.fit_transform(inputs.Sex)
inputs

# OR

# inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
# inputs
```

```python
inputs_n = inputs.drop("Sex", axis="columns")
inputs_n
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, train_size=0.8)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

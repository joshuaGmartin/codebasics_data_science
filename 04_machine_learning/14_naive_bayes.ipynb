# Machine Learning Tutorial Python - 14: Naive Bayes Classifier Algorithm Part 1

- naive = assumption that parameters are independent of each other
- bayes formula uses other know probabilities to calculate the desired unknown probability


```python
import pandas as pd

df = pd.read_csv("data/titanic (1).csv")
df.head()
```

```python
# df.columns
df.drop(
    [
        "PassengerId",
        "Name",
        "SibSp",
        "Parch",
        "Ticket",
        "Cabin",
        "Embarked",
    ],
    axis="columns",
    inplace=True,
)
df.head()
```

```python
target = df.Survived
inputs = df.drop("Survived", axis="columns")
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs = pd.concat([inputs, dummies], axis="columns")
inputs.drop(["Sex", "female"], axis="columns", inplace=True)
inputs.head()
```

```python
inputs.Age.head(10)  # has NaN
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.Age.head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_predict = model.predict(X_train)
y_predict[:10]
```

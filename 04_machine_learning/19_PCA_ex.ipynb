```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/heart.csv")
df.head()
```

```python
df_dumm = pd.get_dummies(df, drop_first=True)
df_dumm.head()
```

```python
z_scores = np.abs((df_dumm - df_dumm.mean()) / df_dumm.std())  # get z=scores
z_scores.head()
```

```python
df_no_outliers = df_dumm[
    (z_scores < 3).all(axis=1)
]  # keep row if all values are within sigma +/-3
df_no_outliers.head()
```

```python
# split
X = df_no_outliers.drop("HeartDisease", axis="columns")
y = df_no_outliers.HeartDisease
X[:5]
```

```python
# scale first
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)

X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, train_size=0.8, random_state=30
)

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

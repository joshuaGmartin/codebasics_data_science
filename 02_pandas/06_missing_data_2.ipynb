## Python Pandas Tutorial 6. Handle Missing Data: replace function


```python
import pandas as pd
import numpy as np

# data has special value for Nan (x: -999999 ). Happends a lot
df = pd.read_csv("data/weather_data (3).csv")
df
```

```python
# df = df.replace([-99999, -88888, 0], np.NaN) # but zero can be valid
new_df = df.replace(
    {
        "temperature": [-99999, -88888],
        "windspeed": [-99999, -88888],
        "event": "0",
    },
    np.NaN,
)  # isolate with dict
new_df
```

```python
new_df = df.replace({-99999: np.NaN, "No Event": "Sunny"})
new_df
```

```python
# new_df = df.replace([" mph", " F|C"], "", regex=True)
new_df = df.replace(
    {
        "temperature": "[A-Za-z]",
        "windspeed": "[A-Za-z]",
    },
    "",
    regex=True,
)
new_df
```

```python
df = pd.DataFrame(
    {
        "student": ["rob", "maya", "parthiv", "tom", "julian", "erica"],
        "score": ["exceptional", "average", "good", "poor", "average", "exceptional"],
    }
)
df
```

```python
# map scores to numbers
new_df = df.replace(["poor", "average", "good", "exceptional"], [1, 2, 3, 4])
new_df
```

## Python Pandas Tutorial 10. Pivot table


```python
# pivot collects common groups and shows data per row, similar to transposing but with extra steps
import pandas as pd
df = pd.read_csv("data/weather.csv")
df
```

```python
# df  = df.pivot(index="date",columns="city")
# df  = df.pivot(index="date",columns="city", values="humidity")
# df = df.pivot(index="humidity", columns="city")
df = df.pivot(index="city", columns="date")
df
```

```python
df = pd.read_csv("data/weather2.csv")
df
```

```python
# df.pivot_table(index="city", columns="date") # auto calcs avg for duplicate data
# df.pivot_table(index="city", columns="date", aggfunc="sum") # sum vs avg
df.pivot_table(index="city", columns="date", margins=True) # "total afffunc" mean of means, sum of sums, etc
```

```python
df = pd.read_csv("data/weather3.csv")
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df.pivot_table(index=pd.Grouper(freq="M", key="date"), columns="city")
```

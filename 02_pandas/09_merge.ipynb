## Python Pandas Tutorial 9. Merge Dataframes


```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "houston"],
    "temperature": [21,14,35, 71],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san francisco"],
    "humidity": [65,68,75],
})
df2
```

```python
# df = pd.merge(df1,df2,on="city") # matches city, not index. But its an intersection/union (exclude non-common)
# df = pd.merge(df1,df2,on="city", how="outer") # includes all
# df = pd.merge(df1,df2,on="city", how="left") # left/right = 1st, 2nd args
# df = pd.merge(df1,df2,on="city", how="right") # left/right = 1st, 2nd args
df = pd.merge(df1,df2,on="city", how="outer", indicator=True) 
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
# df = pd.merge(df1, df2, on="city") # duplicate data for temp and humid
df = pd.merge(df1, df2, on="city", suffixes=("_left", "_right")) 
df
```
